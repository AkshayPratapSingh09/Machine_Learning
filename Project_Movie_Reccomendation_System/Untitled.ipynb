```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
#We will  now go through each column and check which ones are useful to us

#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[["id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head(1)
```

```python
#we will now make 3 columns Movie Id |  title | keywords
#Keywords will be made by merging overview, genres, keywords, cast(main 3 lead) and director's name from Crew
#We will now process data from each columns to merge them together

```

```python
#Remove all the missing values
movies.isnull().sum()
```

```python
#Only 3 entries--> not a major number to remove
movies.dropna(inplace=True)
```

```python
#Check if we have duplicate entries
movies.duplicated().sum()
```

```python
movies.info()
```

```python
import ast
# ast.literal_eval()
```

```python
def convert(obj):
    l = []
    for  i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
    
```

```python
movies.iloc[0].genres
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head(1)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
def convert3(obj):
    l = []
    counter = 0
    for  i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i["name"])
            counter +=1
        else:
            break
        
    return l
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
def fetch_director(obj):
    l = []
    for  i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
# movies["crew_y"] = movies["crew_y"].apply(convert)
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies["overview"][0]
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x :[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x :[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x :[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[["id","title","tags"]]
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df["tags"][0]
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem("dance")
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
#we will have to sort the similarity matrix.
#We will pick 5 movies with least distance
```

```python
# we now search for index position of the movies
new_df[new_df["title"] == "Avatar"].index[0]
```

```python
# testing the movies index search code
new_df[new_df["title"] == "Batman Begins"].index[0]
```

```python
def recommend(movie):
#     new_df[new_df["title"] == "Batman Begins"].index[0]
    movie_index = new_df[new_df["title"] == movie].index[0]
#     print(movie_index)
    
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
